```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'datasets/merged.csv')
drop_cols = ['Student Population','Students to Staff Ratio', 'Times Name', 'QS Name',
             'International Students','Country_QS','Rank_Times', 'Rank_QS','CS', 'CF', 'RI', 'Age', 'STA']
df = df.drop(columns=drop_cols)
```

```python
df.columns
```

```python
score_columns=['Teaching','Research Environment', 'Research Quality', 
               'Industry Impact','International Outlook','AR_SCORE', 
               'ER_SCORE', 'FS_SCORE', 'CPF_SCORE','IF_SCORE', 
               'IS_SCORE', 'IRN_SCORE', 'EO_SCORE', 'S_SCORE', 'Overall_QS']

plt.figure(figsize=(18, 20))
for i, col in enumerate(score_columns, 1):
    plt.subplot(5, 3, i)
    
    sns.histplot(df[col], kde=True, bins=20, color='blue', 
                edgecolor='white', linewidth=0.5)
    
    median = df[col].median()
    mean = df[col].mean()
    median_line = plt.axvline(median, color='black', linestyle='--', linewidth=1.8, label=f'Median: {median:.1f}')
    mean_line = plt.axvline(mean, color='darkred', linestyle='-', linewidth=1.8, label=f'Mean: {mean:.1f}')
    
    plt.title(f'{col}', fontsize=20, pad=10)
    plt.xlabel('Score Value', fontsize=16)
    plt.ylabel('Frequency', fontsize=16)
    plt.legend(fontsize=16)
    plt.xticks(fontsize=12)

plt.tight_layout(pad=2.5)
plt.show()
```

```python
# Box plot comparision
plt.figure(figsize=(10, 5))
sns.boxplot(df[score_columns], palette='viridis')
plt.xticks(rotation=75,fontsize=11)
plt.show()
```
